```python
# check which variations there are between images
# (size, aspect ratio, and color mode)
# This helps ensure consistency before feeding data into a CNN

data_directory = Path(dataset_path)

# Dictionary to store metadata for each image
img_meta = {}

# Loop through each class folder (e.g., Monet, Vangogh, etc.)
for class_folder in Path(dataset_path).iterdir():

    # Ensure we ignore files, only look at folders (each represents a class)
    if class_folder.is_dir():

        # Get all items inside the class folder
        folder_content = class_folder.glob("*")

        # Filter only valid image files:
        # must be a file (not a folder)
        # must be a .jpg image
        images = [img for img in folder_content 
                  if img.is_file() and img.suffix.lower() == ".jpg"]

        # Loop through each image file
        for img_path in images:

            # Open image using PIL (Python Imaging Library)
            with Image.open(img_path) as img:

                # Extract image properties
                width, height = img.size                 # image dimensions
                aspect_ratio = round(width / height, 2)  # shape of the image
                mode = img.mode                          # color format (e.g., RGB, L)

                # Store metadata in dictionary using file name as key    
                img_meta[str(img_path.name)] = {'Width': width,
                                                'Height': height,
                                                'AspectRatio': aspect_ratio,
                                                'Mode': mode}
    
# Convert dictionary into a pandas DataFrame
img_meta_df = pd.DataFrame.from_dict(img_meta, orient='index').reset_index()

# Rename index column to 'FileName' for clarity
img_meta_df.rename(columns={'index': 'FileName'}, inplace=True)

img_meta_df.head()
```

```python
# ── 1. Load ──────────────────────────────────────────────────────────────────
#sdataset_path = "../data/wikiart"
tf.keras.utils.set_random_seed(13)

train_ds = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(128, 128),
    batch_size=32
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(128, 128),
    batch_size=32
)

# ── 2. Extract class info (before any transforms) ────────────────────────────
class_names = train_ds.class_names
num_classes  = len(class_names)

# ── 3. Define transforms ──────────────────────────────────────────────────────
normalization_layer = tf.keras.layers.Rescaling(1./255)

augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomTranslation(0.1, 0.1),
    tf.keras.layers.RandomContrast(0.2),
], name="augmentation")

# ── 4. Apply transforms ───────────────────────────────────────────────────────
AUTOTUNE = tf.data.AUTOTUNE

train_ds = (train_ds
    .map(lambda x, y: (normalization_layer(x), y),        # normalize once
         num_parallel_calls=AUTOTUNE)
    .map(lambda x, y: (augmentation(x, training=True), y), # augment only train
         num_parallel_calls=AUTOTUNE)
    .prefetch(AUTOTUNE))

val_ds = (val_ds
    .map(lambda x, y: (normalization_layer(x), y),        # normalize once
         num_parallel_calls=AUTOTUNE)
    .prefetch(AUTOTUNE))                                   # no augmentation
```
